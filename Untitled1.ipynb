```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor
```

```python
df = pd.read_csv("/content/train.csv")
df.head()
```

```python
df = df[(df["store_nbr"] == 1) & (df["family"] == "AUTOMOTIVE")]
df = df[["date", "sales"]]
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")
df.head()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df["date"], df["sales"])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()

```

```python
df["lag_1"] = df["sales"].shift(1)
df["lag_7"] = df["sales"].shift(7)
df["lag_30"] = df["sales"].shift(30)

df["rolling_7"] = df["sales"].rolling(7).mean()
df["rolling_14"] = df["sales"].rolling(14).mean()

df["day"] = df["date"].dt.day
df["month"] = df["date"].dt.month
df["dayofweek"] = df["date"].dt.dayofweek

df.dropna(inplace=True)
df.head()
```

```python
train_size = int(len(df) * 0.8)
train = df.iloc[:train_size]
test = df.iloc[train_size:]

X_train = train.drop(["date", "sales"], axis=1)
y_train = train["sales"]

X_test = test.drop(["date", "sales"], axis=1)
y_test = test["sales"]
```

```python
xg_model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

xg_model.fit(X_train, y_train)
```

```python
xg_preds = xg_model.predict(X_test)

mae = mean_absolute_error(y_test, xg_preds)
rmse = np.sqrt(mean_squared_error(y_test, xg_preds))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
plt.figure(figsize=(12,4))
plt.plot(test["date"], y_test.values, label="Actual")
plt.plot(test["date"], xg_preds, label="Predicted")
plt.legend()
plt.title("Actual vs Predicted Sales")
plt.show()
```

```python
last_row = df.iloc[-1:].drop(["date", "sales"], axis=1)

future_preds = []
current = last_row.copy()

for _ in range(14):
    pred = xg_model.predict(current)[0]
    future_preds.append(pred)

    current["lag_30"] = current["lag_7"]
    current["lag_7"] = current["lag_1"]
    current["lag_1"] = pred
    current["rolling_7"] = np.mean(future_preds[-7:])
    current["rolling_14"] = np.mean(future_preds[-14:])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df = df[(df["store_nbr"] == 1) & (df["family"] == "AUTOMOTIVE")]
df = df[["date", "sales"]]

df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")
df.reset_index(drop=True, inplace=True)
```

```python
df["lag_1"] = df["sales"].shift(1)
df["lag_7"] = df["sales"].shift(7)
df["lag_14"] = df["sales"].shift(14)
df["lag_30"] = df["sales"].shift(30)

df["rolling_7"] = df["sales"].rolling(7).mean()
df["rolling_14"] = df["sales"].rolling(14).mean()
df["rolling_30"] = df["sales"].rolling(30).mean()

df["day"] = df["date"].dt.day
df["month"] = df["date"].dt.month
df["dayofweek"] = df["date"].dt.dayofweek

df.dropna(inplace=True)
df.head()
```

```python
train_size = int(len(df) * 0.8)

train = df.iloc[:train_size]
test = df.iloc[train_size:]

X_train = train.drop(["date", "sales"], axis=1)
y_train = train["sales"]

X_test = test.drop(["date", "sales"], axis=1)
y_test = test["sales"]
```

```python
X_scaler = StandardScaler()

X_train_scaled = X_scaler.fit_transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
y_scaler = StandardScaler()

y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1,1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1,1))
```

```python
model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.1),
    Dense(1)
])

optimizer = Adam(learning_rate=0.0005)

model.compile(
    optimizer=optimizer,
    loss=Huber()
)

model.summary()
```

```python
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = model.fit(
    X_train_scaled,
    y_train_scaled,
    validation_split=0.1,
    epochs=200,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)
```

```python
preds_scaled = model.predict(X_test_scaled)
preds = y_scaler.inverse_transform(preds_scaled).flatten()
```

```python
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))

print("Dense NN MAE:", mae)
print("Dense NN RMSE:", rmse)
print("Mean Sales:", y_test.mean())
```

```python
plt.figure(figsize=(12,4))
plt.plot(y_test.values, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("Dense Neural Network  Actual vs Predicted")
plt.show()
```

```python
errors = y_test.values - preds

plt.hist(errors, bins=30)
plt.title("Prediction Error Distribution")
plt.show()
```
